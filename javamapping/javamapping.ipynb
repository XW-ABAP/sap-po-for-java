1) NWDS软件概述及其与SAP PI的关系
NWDS 是Net Weaver Developer Studio的缩写。它是一个基于 Eclipse的集成开发环境( IDE)，可用于编写 SAP 的 Java 应用程序（Eclipse 工具经过修改，专为 SAP 而设计）。对于 PI 相关功能，您可以使用 NWDS 创建数据类型/消息类型、各种类型的映射、创建自定义适配器模块等等。

NWDS 还允许您创建和定义 Web 服务。

总结一下NWDS的用途：

编写独立的Java程序
编写和编译 Java 映射
编写、编译和部署适配器模块
连接到 SAP PI 系统
编写目录 API 以连接到 SAP PI 系统
创建、配置和部署 NW BPM
SAP PI 开发工具
开发与 iFlows 的接口
2)检查 SAP PI AEX JDK 版本
在安装 NWDS 并开始配置之前，我们先来确定 PI 系统的 JDK 版本。例如，如果您是项目新手，客户提供了 PI Java Stack 系统 URL，那么您需要在系统中安装 JDK 才能访问该系统，否则您将无法打开 ESR/ID。

PI JDK 版本应与您的系统 JDK 版本匹配。要了解 PI JDK 版本：访问 PI 主页 -> http://<host>:<port>/dir/start/index.jsp，然后单击下方所示的“管理”链接，打开管理工具窗口。


点击“软件版本信息”（ESR 或 ID 选项卡），即可查看其中显示的 JDK 版本。因此，您需要将机器上的 JDK 版本与 PI 系统上的 JDK 版本进行匹配。


3)了解 JDK 和 JRE 之间的区别。
JDK – Java 开发工具包；
JRE – Java 运行时环境

简单来说，系统需要 JRE 才能运行 Java 程序，例如运行/打开 ESR 和 ID。JRE 在您只需要运行 Java 程序时非常有用。如果没有 JRE，开发人员将无法开发 Java 程序。

开发者需要 JDK 来开发 Java 程序。例如，如果你想编写 Java 程序、编译并执行它，就需要 JDK。如今，JDK 已经包含 JRE，所以你无需单独考虑 JRE，只需 JDK 即可。

4)在本地计算机上下载并配置 Java
https://www.oracle.com/in/java/technologies/javase-downloads.html

访问上述网址并下载 JAVA SE 8 JDK。安装它。


安装完成后，打开命令行并核实已安装的版本。输入“java -version”并按回车键即可查看已安装的版本。


已安装的版本与 PI JDK 版本匹配。因此，我们的方向正确。要查看本地计算机上的 JDK 文件，请导航至“C:\Program Files\Java”。

如果您在命令行中看不到 JAVA 版本详细信息，请不要担心，一旦您在环境变量中配置了 java 文件路径 (C:\Program Files\Java\jdk1.8.0_291\bin)，它就会显示出来。

要设置上述路径，让我们进入系统属性。

右键单击“此电脑”或“我的电脑”
选择属性
点击高级系统设置
在“高级”选项卡中，单击“环境变量”按钮
环境变量屏幕弹出
在“系统变量”部分，应该有一个 PATH 变量。
点击它，然后选择“编辑”。
点击“新建”按钮，添加上述“bin”路径。
现在在命令行中，输入“java -version”并按回车键显示 JDK 版本详细信息。
5)下载并安装 NWDS
下载 NWDS 软件所需详细信息。SAP PI 系统的 SP 和补丁级别，您可以在 NWA 的系统组件信息部分找到。NWA -> 配置选项卡 -> 基础架构子选项卡 -> 系统信息链接。

在“组件信息”选项卡下，筛选出名称列中的 *XIAF*。


对应的版本信息可以在“版本”栏中查看。

1000.7.50.10.78.20200331203000
1000 – 不相关
7.50 – 这是 PI 系统的 NetWeaver 版本
10 – 这是系统的 SP 版本
78 – 这是 SAP_XIAF 组件的补丁级别
20200331203000 – 这是补丁发布的日期/时间，在本例中为 2020/03/31 20:30:00

在此版本中，XIAF 组件位于 SP10 补丁 78 中。

您可以忽略补丁级别。SP 版本对于下载 NWDS 至关重要。请访问以下 URL：SAP 软件中心。

https://launchpad.support.sap.com/#/softwarecenter

您需要一个S用户ID才能从SAP软件中心下载软件。


导航至以下路径：按字母索引 (AZ) -> N -> SAP NetWeaver -> SAP NetWeaver 7.5 -> 点击右上角的蓝色“支持包和补丁”按钮 -> 开发者工作室 -> NW 开发者工作室 7.5

对于我的 SP 等级 10，我下载了 NWDS SP10 P18 zip 文件，大小约为 1.1GB。


下载完成后，您可以看到生成的 ZIP 文件。

最重要的一步是：请使用 7-Zip 解压，不要使用 WinRAR。使用 WinRAR 解压时似乎会出现一些问题，例如文件损坏或缺少某些文件，导致解压失败。因此，请使用 7-Zip 来解决这个问题。

解压后，双击 ECLIPSE.exe 文件运行应用程序。


应用程序会提示我们输入工作区名称和文件夹路径（项目文件位置）。工作区位置可以为每个项目定义，既可以在启动 NWDS 时定义，也可以在之后手动定义。输入并点击“继续”后，应用程序会抛出一个错误，提示“您正在使用不受支持的 JVM 运行 Developer Studio。您必须使用 JDK 作为 JVM，而不是 JRE ”。

6)安装 NWDS 时遇到的错误
错误信息显示，我们安装的 JDK 无法被 NWDS 软件识别。因此，我们需要指定 NWDS 使用此版本的 JDK。

NWDS中配置JDK路径的两种方法：

a) 在 NWDS 软件中 -> Windows 菜单 -> 首选项 -> Java 下拉菜单 -> 已安装的 JRE。您需要在此处添加 JDK 路径，方法是单击如下所示的“添加”按钮。




这种方法也有其缺点，即需要在“首选项”中手动输入 JDK 路径，因为每次创建工作区时，都需要在“窗口”->“首选项”菜单中配置 JDK 路径。不过，如果您希望某些项目使用旧版本的 JDK，而另一些项目使用新版本的 JDK，那么这种方法就很有用了。

b) 否则，使用此方法在启动 NWDS 软件时通过更改 ECLIPSE.INI 文件中的启动参数来配置 JDK 路径。



用记事本打开 ECLIPSE.INI 文件，并添加以下参数

-vm

C:/Program Files/Java/jdk1.8.0_291/bin/javaw.exe




好了，现在您可以自由使用NWDS（SAP增强版的Eclipse）了。

7)将NWDS与SAP PI系统连接
现在我们将把PI系统连接到NWDS，并尝试从NWDS访问ESR和ID内容。在开始连接过程之前，您需要知道PI服务器的主机和端口详细信息。

导航至“窗口”->“首选项”->“SAP AS JAVA”->单击“添加”以添加新连接。







连接详细信息维护完毕后，点击“连接”连接到 PI ESR 和 ID，然后输入 PI 用户凭据。






选择“对目录和存储库使用通用凭据”，然后继续连接。

8）了解NWDS环境
连接成功后，您将能够在左侧窗格中看到企业服务存储库视图。如果看不到，请单击“窗口”->“透视图”->“打开透视图”->“其他”，然后选择“企业服务存储库”。






要显示集成目录对象，同样地，打开透视图并选择 SAP 流程集成设计器。



在“集成流和系统”下拉列表中，您将看到已创建的 IFlows 列表，其中包含已创建的业务系统和业务组件及其各自的通信渠道。



要在 ESR 视图和 ID 视图之间切换，请单击相应的选项卡，或选择“快速访问”旁边的按钮，如下所示。



菜单选项“流程集成”可用于快速访问，例如查看值映射组/通过 CTS 或导出/导入传输对象。此外，您只需单击一个按钮，即可通过选择“流程集成”->“生成集成流”->“集成配置”将批量集成方案转换为iFlow。
